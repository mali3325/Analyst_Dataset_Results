```python
# Question 5
ROAS_by_audience = merged_data[['Audience Type', 'Audience', 'Media Spend', 'Total Revenue']]
ROAS_by_audience = ROAS_by_audience.set_index(['Audience Type', 'Audience'])
ROAS_by_audience = ROAS_by_audience.groupby('Audience').sum()
ROAS_by_audience['ROAS (in %)'] = ((ROAS_by_audience['Total Revenue'] / ROAS_by_audience['Media Spend'])*100) - 100
ROAS_by_audience = ROAS_by_audience.sort_values('ROAS (in %)')

def highlight_negative(s):
    '''
    highlight the negative in a Series red.
    '''
    is_negative = s <= 0
    return ['background-color: red' if v else '' for v in is_negative]

ROAS_by_audience.style.apply(highlight_negative)

# the red cell below indicates that Return on ad spent is negative (66.23% - 100%) = -33.76%) 
# thus targeting Facebook Interest audience interested in competitors brought in a loss of 2287.8 
# the audience should be dropped immediately
```

```python
# the table below shows that Facebook Interest audience interested in competitors 
# resulted in the smallest number of likes out of all categories in the audience. 
# and the number of shares for each audience might suggest that homepage audience which falls under 
# website custom audience type tend to share the content more than other target audience groups. 
 
def highlight_min(s):
    '''
    highlight the min in a Series red.
    '''
    is_min = s == s.min()
    return ['background-color: red' if v else '' for v in is_min]

merged_data[['Audience', 'Likes', 'Shares', 'Purchases']].groupby('Audience').sum().style.apply(highlight_min)
```

```python
audience_table = merged_data[['Optimization Event','Audience Type','Audience', 'Media Spend', 'Total Revenue']]
audience_table = audience_table.set_index(['Optimization Event','Audience Type', 'Audience'])
# only for 'Conversion' Optimization Event 
ROAS_by_conversion = audience_table.loc['Conversion'].groupby('Audience').sum()
ROAS_by_conversion['ROAS (in %)'] = ((ROAS_by_conversion['Total Revenue'] / ROAS_by_conversion['Media Spend'])*100) - 100
ROAS_by_conversion.loc['Total'] = [ROAS_by_conversion['Media Spend'].sum(), 
                                   ROAS_by_conversion['Total Revenue'].sum(), 
                                 ((ROAS_by_conversion['Total Revenue'].sum()/ROAS_by_conversion['Media Spend'].sum())*100) -100 ] # adding a row
def highlight_last_row(s):
    return ['background-color: green' if i==len(s)-1 else '' for i in range(len(s))]
ROAS_by_conversion.style.apply(highlight_last_row)
# the table below shows ROAS by 'Conversion' relating to each audience. 
# as you will see from the next table, 'Conversion' optimization event yields a higher profit compared to 'Clicks'
```

```python
ROAS_by_click = audience_table.loc['Click'].groupby('Audience').sum()
ROAS_by_click['ROAS (in %)'] = (((ROAS_by_click['Total Revenue'] / ROAS_by_click['Media Spend'])*100) - 100)
ROAS_by_click.loc['Total'] = [ROAS_by_click['Media Spend'].sum(), 
                                   ROAS_by_click['Total Revenue'].sum(), 
                                   ((ROAS_by_click['Total Revenue'].sum()/ROAS_by_click['Media Spend'].sum())*100) - 100] # adding a row
def highlight_last_row(s):
    return ['background-color: green' if i==len(s)-1 else '' for i in range(len(s))]
ROAS_by_click.style.apply(highlight_last_row)
```
